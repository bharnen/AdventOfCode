```python

class CircuitHolder:
    
    def __init__(self, data):
        self.programs = []
        self.stucture = self.parseData(data)
        
    
    def getProgram(self, name):
        for i in range(0, len(self.programs)):
            current_program = self.programs[i]
            if current_program.name == name:
                return current_program
        new_prog = Program(name)
        self.programs.append(new_prog)
        return new_prog
        
    
    def parseData(self,data):
        lines = self.getLines(data)
        for line in lines:
            self.parseLine(line)
        
        
    def parseLine(self, line):
        line_words = line.split()
        program_name = line_words[0]
        program_weight = line_words[1]
        program_weight = int(self.sanitize(program_weight))
        
        prog = self.getProgram(program_name)
        prog.weight = int(program_weight)  
        if len(line_words) != 2:
            holding = self.getHoldingPrograms(line_words)
            prog.holding = holding
            self.handlePreviousAssignment(prog.name, holding)


        
    def getHoldingPrograms(self, line_words):
        if len(line_words) <= 2:
            return None
        else:
            holding_programs = []
            for i in range (0, len(line_words)-3):
                prog = line_words[3+i]
                prog = self.sanitize(prog)
                holding_programs.append(prog)
        return holding_programs
                
        
    def handlePreviousAssignment(self, main, holding):
        for name in holding:
            held_program = self.getProgram(name)
            held_program.previous = main
            
    
        
    def getLines(self, data):
        lines = []
        line = ''
        for c in data:
            if c == '\n':
                lines.append(line)
                line = ''
            else:
                line += c
        if line != '':
            lines.append(line)
            line = ''
        return lines
    
    def sanitize(self, word):
        word = word.strip(' ')
        new_word = ''
        for c in word:
            if c != ',' and c != '(' and c != ')':
                new_word += c
        return new_word
    
    def baseProgram(self):
        for i in range(0, len(self.programs)):
            current_program = self.programs[i]
            if current_program.previous is None:
                return current_program.name
            
    def toString(self):
        for i in range(0, len(self.programs)):
            current_program = self.programs[i]
            print("Program: " + str(current_program.name) + " weight: " + str(current_program.weight) + " previous:"
             + str(current_program.previous) + " holding: " + str(current_program.holding))

        
    def checkBalance(self):
        base_program = self.getProgram(self.baseProgram())
        print("Base Program: " + str(base_program.name) + " holding: " + str(base_program.holding))
        for p in base_program.holding:
            current_program = self.getProgram(p)
            print("prog: " + str(current_program.name) + " weight: " + str(self.getWeightSum(current_program)))
        
    
    def getWeightSum(self, program):
        if len(program.holding) > 0:
            sum = 0
            for p in program.holding:
                p = self.getProgram(p)
                added_weight = self.getWeightSum(p)
                sum += added_weight
            return int(program.weight) + sum
        else:
            return program.weight
        
    
    def fixUnbalancedProgram(self, program_name, msg):
        print()
        print()
        p = self.getProgram(program_name)
        for i in range(0, len(p.holding)):
            current_held_program = self.getProgram(p.holding[i])
            weight = self.getWeightSum(p)
            print("prog: " + str(p.name) + " weight: " + str(weight))
            real_msg = msg + str(p.weight) + '+ ('
            for j in range(0, len(p.holding)):
                real_msg += str(self.getWeightSum(self.getProgram(p.holding[j]))) + ' + '
            real_msg += ')'
            print(real_msg)
            self.fixUnbalancedProgram(p.holding[i], msg+'\t')
        
    

```
