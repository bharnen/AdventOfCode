```python


class MancalaPatternFinder:
    
    def __init__(self,data):
        self.list = self.parseData(data)
        self.cache = []
        
        
    def parseData(self,data):
        result = []
        current_value = ''
        for c in data:
            if c == '\t':
                result.append(int(current_value))
                current_value = ''
            else:
                current_value += c
        if c != '':
            result.append(int(current_value))
            current_value = ''
        return result
    
    
    def distributePieces(self):
        maxPieceIndex = self.findMaxPieceIndex()
        maxPieceValue = self.list[maxPieceIndex]
        self.list[maxPieceIndex] = 0
        current_index = maxPieceIndex+1
        if current_index == len(self.list):
            current_index = 0
        while maxPieceValue > 0:
            #print("current: " + str(current_index) + " len: " + str(len(self.list)))
            self.list[current_index] += 1
            maxPieceValue -= 1
            current_index += 1
            if current_index == len(self.list):
                current_index = 0
        
        
    def findMaxPieceIndex(self):
        maximum = self.list[0]
        for i in range(0, len(self.list)):
            current_value = self.list[i]
            if current_value > maximum:
                maximum = current_value
        return self.list.index(maximum)
    
    
    def addToCache(self):
        list_as_string = self.stringify()
        self.cache.append(list_as_string)
        
    def existsInCache(self):
        current_pattern = self.stringify()
        for s in self.cache:
            if current_pattern == s:
                return True
        return False
        
        
    def stringify(self):
        return ''.join(str(e) for e in self.list)
    
    
    def findRepeatedPattern(self):
        mancala_steps = 0
        while self.existsInCache() != True:
            mancala_steps += 1
            self.addToCache()
            self.distributePieces()
        self.addToCache()
        return mancala_steps
    
    
    def infiniteLoopSize(self):
        temp = []
        for i in self.cache:
            temp.append(i)
        repeated_pattern = temp[len(temp)-1]
        del temp[len(temp)-1]
        return len(temp) - temp.index(repeated_pattern)
        

                
    def toString(self):
        print(str(self.list))
        
        
        
        
```

```python
data = '''4	1	15	12	0	9	9	5	5	8	7	3	14	5	12	3'''

a = MancalaPatternFinder(data)
print(str(a.findRepeatedPattern()))
print(str(a.infiniteLoopSize()))

```
